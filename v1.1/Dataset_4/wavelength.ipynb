```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate LED curve
def generate_LED_curve(wavelength_peak, intensity_peak, width=50):
    wavelength = np.arange(500, 2001, 1)
    intensity = np.exp(-((wavelength - wavelength_peak) / width) ** 2)
    intensity = intensity / np.max(intensity) * intensity_peak
    return wavelength, intensity

# Set up LED curves
led_curves = [
    {'peak': 850, 'intensity': 1},
    {'peak': 940, 'intensity': 1},
    {'peak': 1050, 'intensity': 1},
    {'peak': 1200, 'intensity': 1},
    {'peak': 1300, 'intensity': 1},
    {'peak': 1460, 'intensity': 1},
    {'peak': 1550, 'intensity': 1},
    {'peak': 1650, 'intensity': 1},
]

# Plot superimposed wavelengths for adjacent LED curves
for i in range(len(led_curves) - 1):
    led_curve1 = led_curves[i]
    led_curve2 = led_curves[i + 1]

    curve1_wavelength, curve1_intensity = generate_LED_curve(led_curve1['peak'], led_curve1['intensity'])
    curve2_wavelength, curve2_intensity = generate_LED_curve(led_curve2['peak'], led_curve2['intensity'])

    superimposed_intensity = curve1_intensity + curve2_intensity

    plt.plot(curve1_wavelength, superimposed_intensity, label=f"LED {i+1} nm + LED {i+2} nm")

# Set graph limits
plt.xlim(500, 2000)
plt.ylim(0, np.max(superimposed_intensity) + 0.2)  # Adjust ylim based on the intensity range

# Add labels and legend outside the graph
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
plt.title('Superimposed LED Intensity vs. Wavelength')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot or save it to a file
plt.show()
# To save the plot to a file, uncomment the following line
# plt.savefig('superimposed_adjacent_led_intensity_vs_wavelength.png', bbox_inches='tight')
```
